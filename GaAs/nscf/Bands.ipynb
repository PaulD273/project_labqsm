# Bands

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
#%matplotlib widget
```

```python
data = []
with open('bands_gaas.dat.gnu','r') as file :
    csv_reader = csv.reader(file, delimiter=' ', skipinitialspace=True)
    for row in csv_reader:
        if any(row):
            data.append(row)
            
    f=np.asarray(data, dtype='f8')
                        
print(f)
```

```python
fig, ax = plt.subplots()
plt.scatter(f[:,0],f[:,1]-5.1667,marker=".")
plt.scatter(f[:804,0],f[:804,1]-5.1667,marker=".",color='r')
plt.ylabel(r"Energy [eV]")
plt.xlabel(r"BZ path")
ax.set_xticks(ticks=[0, 1, 3.9873, 2.5731, 2.9267], labels = ['L', '$\Gamma$', '$\Gamma$', 'X', 'K'])
plt.axvline(x=1, linewidth=0.5, color='k')
plt.axvline(x=2.5731, linewidth=0.5, color='k')
plt.axvline(x=2.9267, linewidth=0.5, color='k')
plt.axhline(y=0, linewidth=1, color='k', ls='--')
plt.axhline(y=1.351, linewidth=1, color='k', ls='--')

ax.annotate('$\Lambda$',
            xy=(150, 32), xycoords='figure pixels')

ax.annotate('$\Delta$',
            xy=(280, 32), xycoords='figure pixels')

ax.annotate('$\Sigma$',
            xy=(480, 32), xycoords='figure pixels')

#ax.set_ylim(min(f[:,1]),max(f[:,1]))
#ax.set_xlim(0,3.9873)
#plt.legend()
```
