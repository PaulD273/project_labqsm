```python
f=read_dos_file.read_dos_file('gaas_025.dat')
print(f)
```

```python
f1=read_dos_file.read_dos_file('gaas_01.dat')
print(f1)
```

```python
fig, ax = plt.subplots()
plt.scatter(f[:,0],f[:,1],marker=".", label='0.025 Ry')
plt.scatter(f1[:,0],f1[:,1],marker=".", label='0.01 Ry')
plt.xlabel(r"Energy [eV]")
plt.ylabel(r"DOS [#$e^{-}$/eV]")
plt.axvline(x=6.5878, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
plt.axvline(x=6.7245, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
plt.fill_between(f[:,0], 0, f[:,1], where=(f[:,0] < 6.206), facecolor='red', alpha=0.25)
plt.text(5.6, 1.5, 'Energy Gap', rotation=90)
plt.legend()
```
