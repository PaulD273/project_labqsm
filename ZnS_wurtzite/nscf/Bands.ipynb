```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
%matplotlib widget
```

```python
data = []
with open('bands_zns_w_2.dat.gnu','r') as file :
    csv_reader = csv.reader(file, delimiter=' ', skipinitialspace=True)
    for row in csv_reader:
        if any(row):
            data.append(row)
            
    f=np.asarray(data, dtype='f8')
                        
print(f)
```

```python
fig, ax = plt.subplots()
#plt.scatter(f[:,0],f[:,1]-4.7848,marker=".")
#plt.scatter(f[:2613,0],f[:2613,1],marker=".",color='r')
plt.scatter(f[:,0],f[:,1],marker=".",color='b')
plt.ylabel(r"Energy [eV]")
plt.xlabel(r"BZ path")
ax.set_xticks(ticks=[0, 0.6667, 1.2440, 1.8214], labels = ['K', '$\Gamma$', 'M', '$\Gamma$'])
#plt.axvline(x=1, linewidth=0.5, color='k')
#plt.axvline(x=2.5731, linewidth=0.5, color='k')
#plt.axvline(x=78.9371, linewidth=0.5, color='k')
plt.axhline(y=83.7365, linewidth=1, color='k', ls='--')
plt.axhline(y=78.9371, linewidth=1, color='k', ls='--')
plt.axhline(y=77.0182, linewidth=1, color='k', ls='--')


ax.set_ylim(70,90)
#ax.set_ylim(0,14)
#ax.set_xlim(0,3.9873)
#plt.savefig('bands_zns_w_2.pdf')
#plt.legend()
```
